```python
sales_by_city = df.groupby('city')['sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(10,5))
sales_by_city.plot(kind='bar')
plt.title("Sales by City")
plt.xlabel("City")
plt.ylabel("Sales in USD")
plt.show()
```

```python
sales_by_month = df.groupby('month')['sales'].sum()
months = range(1, 13)

plt.figure(figsize=(10,5))
plt.bar(months, sales_by_month)
plt.xticks(months)
plt.xlabel("Month")
plt.ylabel("Sales in USD")
plt.title("Total Sales by Month")
plt.show()
```

```python
orders_by_hour = df.groupby('hour')['order_id'].count()

plt.plot(orders_by_hour.index, orders_by_hour.values)
plt.title("Number of Orders by Hour")
plt.xlabel("Hour of the Day")
plt.ylabel("Number of Orders")
plt.grid(True)
plt.show()
```
